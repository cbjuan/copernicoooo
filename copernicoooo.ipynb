```python
import requests
from datetime import datetime
from bokeh.io import push_notebook, show, output_notebook
from bokeh.plotting import figure
from bokeh.models import (
    HoverTool,
    ColumnDataSource
)
output_notebook()
```

```python
r = requests.get('https://wikimedia.org/api/rest_v1/metrics/pageviews/per-article/es.wikipedia/all-access/all-agents/Nicol%C3%A1s_Cop%C3%A9rnico/daily/2017010100/2017011400')
```

```python
#print r
#print r.text
r.encoding
results = r.json()
```

```python
x = []
sX = []
y = []
for result in results['items']:
    dia = datetime.strptime(((result['timestamp'])[0:8]), '%Y%m%d').date()
    visitas = result['views']
    articulo = result['article']
    sDia = str(dia)
    x.append(dia)
    sX.append(str(sDia))
    y.append(visitas)
```

```python
incidencia = []
for s in sX:
    #print s
    if s == '2017-01-05':
        incidencia.append("OKjeda hace el vídeo")
    elif s == '2017-01-06':
        incidencia.append("OKDiario publica la noticia")
    elif s == '2017-01-07':
        incidencia.append("Twitter Explota -> Soy Copernicooo")
    else:
        incidencia.append("-")
```

```python
p = figure(title="Visitas a la página 'Nicolás Copérnico' en Wikipedia ES", x_axis_type="datetime", tools=[])
sourcetds = ColumnDataSource(
    data=dict(
        dia  = x,
        visitas = y,
        sDia = sX,
        incidencias = incidencia
    )
)

p.line(x='dia', y='visitas', legend="Nº visitas", source = sourcetds, line_width=2)
p.circle(x='dia', y='visitas', source = sourcetds)

hover = HoverTool(
    tooltips=[
        ("Fecha", "@sDia"),
        ("Nº visitas", "@visitas"),
        ("¿Qué ocurrió", "@incidencias")
    ]
)
p.add_tools(hover)
show(p)
```
